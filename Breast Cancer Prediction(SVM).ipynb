```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
result = load_breast_cancer()
result
```

```python
result.feature_names
```

```python
result.keys()
```

```python
print(result['data'].shape)
```

```python
df_data = pd.DataFrame(data=np.c_[result['data'],result['target']], columns=np.append(result['feature_names'],['target']))
```

```python
df_data
```

```python
sns.pairplot(data=df_data, hue='target', height=3, vars=['mean radius','mean texture','mean symmetry','mean area'])
```

```python
sns.countplot(df_data['target'])
```

```python
plt.figure(figsize=(20,14))
sns.heatmap(df_data.corr(), annot=True)
```

```python
X = df_data.drop(['target'], axis=1)
y=df_data['target']
```

```python
X
```

```python
y
```

```python
X_train, x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=5 )
```

```python
X_train
```

```python
x_test
```

```python
y_test.head(20)
```

```python
from sklearn import svm
```

```python
model = svm.SVC()
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test, y_predict) 
cm
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
model.score(x_test,y_test)
```
